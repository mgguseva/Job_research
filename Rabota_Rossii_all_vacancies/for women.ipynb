```python
import pandas as pd

df = pd.read_csv("работа россии.csv", parse_dates=["vacancy.creation-date"], infer_datetime_format=True)
```

```python
df
```

```python
plot = df[df["vacancy.creation-date"] > "2021-12-31"]
plot["vacancy.creation-date"] = plot["vacancy.creation-date"].dt.to_period("M")
```

```python
plot
```

```python
specialisations = pd.pivot_table(plot, index = ['vacancy.creation-date', "vacancy.category.specialisation"], aggfunc='count')
```

```python
specialisations = specialisations.rename(columns={"Unnamed: 0": "value"})
```

```python
import altair as alt

alt.Chart(df).mark_area().encode(
    x="vacancy.creation-date:T",
    y="sum(value):Q",
    color="vacancy.category.specialisation:N"
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()

plt.stackplot(plot["vacancy.creation-date"], plot["vacancy.category.specialisation"])
              # labels=['Team A', 'Team B', 'Team C'],
              # colors=color_map)

#add legend
plt.legend(loc='upper left')

#add axis labels
plt.xlabel('Period')
plt.ylabel('Points Scored')

#display area chart
plt.show()
```
